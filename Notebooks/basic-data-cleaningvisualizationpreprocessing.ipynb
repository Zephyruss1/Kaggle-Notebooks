```python
def show_null_data(data):
    null_count   = data.isna().sum()
    null_percent = data.isna().sum() / len(data)
    null_percent = null_percent.apply(lambda x: f"{x:.1%}")
    print("Total Duplicated Data:", data.duplicated().sum())
    result = pd.concat([null_count, null_percent], axis=1)
    result.columns = ['Null Data Count', 'Null Percentage']
    return result
show_null_data(df)
```

### Data Cleaning and PREPROCESSING

```python
df.info()
```

#### **List comprehension:**

#### Civillian aircraft --> 0. 
#### Military aircraft --> 1.

```python
df['C_M'] = [0 if i == 'C' else 1 for i in df.C_M]
```

```python
df
```

```python
df['Type_Role'] = [i.replace('/', '_').replace('\xa0', ' ').replace(' ', '').strip() for i in df['Type_Role']]
```

```python
df['Type_Role'].unique()
```

```python
df_dm = df[['Type_Role', 'Country']]
```

```python
df_dm = pd.get_dummies(df_dm)
```

```python
df_dm
```

```python
prep_df = pd.concat([df, df_dm], axis=1)
```

```python
prep_df.drop(['Type_Role', 'Notes', 'Name', 'Country'], axis=1, inplace=True)
```

```python
prep_df
```

```python
prep_df['Number_produced'] = [i.replace('+', ' ').replace('[',' ').replace(']', ' ').replace(',', '').strip() for i in prep_df['Number_produced']]
```

```python
prep_df['Number_produced'] = [i.split(' ')[0] for i in prep_df['Number_produced']]
```

```python
prep_df['Number_produced']
```

```python
import re

def clean_numeric(value):
    # If there is NaN value return 'value'.
    if pd.isna(value):
        return value
    # Clean the characthers instance is non numeric 
    cleaned_value = re.sub(r"[^0-9]", "", str(value))
    return cleaned_value

prep_df['Number_produced'] = prep_df['Number_produced'].apply(clean_numeric)
```

```python
prep_df['Number_produced'] = prep_df['Number_produced'].astype('int64')
```

```python
prep_df.info()
```

```python
prep_df
```

```python
prep_df['End'].unique()
```

```python
prep_df['End'] = [2024 if i == 'present' else i for i in prep_df['End']]
```

```python
prep_df['End'] = prep_df['End'].astype('str')
```

```python
prep_df['End']
```

```python
prep_df['End'] = [i.replace('[','').replace(']','').replace('s', '').strip() for i in prep_df.End]
```

```python
prep_df['End'].unique()
```

```python
def equal_len(data):
    new_df = []
    for i in data:
        if len(str(i)) != 4:
            new_df.append(str(i[:4]))
        else:
            new_df.append(str(i))
    return new_df
prep_df['End'] = equal_len(prep_df['End'])
```

```python
prep_df['End']
```

```python
prep_df.End = prep_df.End.astype('int64')
```
